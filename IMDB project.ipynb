__STEPS:__
To query data in an SQLite database from Python, you use these steps:

- First, establish a connection to the SQLite database by creating a Connection object.
- Next, create a Cursor object using the cursor method of the Connection object.
- Then, execute a SELECT statement.
- After that, call the fetchall() method of the cursor object to fetch the data.
- Finally, loop the cursor and process each row individually.

### 7.Find the name of the 10th firstwomen directors

```python

query ="SELECT name from directors where gender like '1' order by id asc limit 1 offset 9 "
cur.execute(query)
#SELECT name: Specifies that you want to retrieve the name column from the directors table.
#FROM directors: Specifies the table from which to retrieve the data, in this case, the directors table.
#WHERE gender = '1': Filters the rows to include only those where the gender column is equal to '1'.
#ORDER BY id ASC: Orders the result set in ascending order based on the id column. The ASC keyword is optional because sorting is in ascending order by default.
#LIMIT 1 OFFSET 9: Limits the result set to 1 row, starting from the 10th row (offset 9). This effectively retrieves the 10th row based on the specified ordering.

result=cur.fetchall()
#fetchall() --> fetch all rows for the current query, if no rows avaliable then,it returns an empty list.
for row in result:
    print('Top 10 th first women :', row[0])
# for -It's commonly used to iterate over a sequence
```

### 10.What is the most awarded average vote since the January 1st,2000

```python
# SQL query
query="SELECT original_title from Movies where release_date >= '2000-01-01' order by vote_average desc limit 1 "
#This SQL query retrieves the original_title from the Movies table, filtering the result to include only rows where the release_date is on or after January 1st, 2000.
#The result is ordered by the vote_average column in descending order, and only the top result is selected using LIMIT 1

# Execute the query
cur.execute(query)

# Fetch all rows from the result set
result =cur.fetchall()

# Iterate over the result set and print the original_title of the most awarded movie
for row in result:
    print("Most awarded average vote Movie since january 1st,2000 is :",row[0])
```

### 11.Which movies were directed by Brenda Chapman

```python
# SQL query
query="SELECT original_title from Movies M inner join directors D on M.director_id=D.id where D.name like 'Brenda Chapman'"
#This SQL query retrieves the original_title from the Movies table, using an inner join with the directors table on the id column. 
#It filters the result to include only rows where the director's name is 'Brenda Chapman'.

# Execute the query
cur.execute(query)

# Fetch all rows from the result set
result=cur.fetchall()

# Iterate over the result set and print the original_title of each movie
for row in result:
    print("Movies Directed by Brenda:",row[0])
```

### 12.Whose director made the most Movies

```python
# SQL query
query="SELECT D.name ,count(M.original_title) as countofmovies from movies M inner join directors D on M.director_id=D.id group by D.name ORDER BY countofmovies DESC limit 1"
#This SQL query retrieves the name of directors and counts the number of movies each director has made from the movies table, using an inner join with the directors table.
#The result is grouped by director names, ordered by the count of movies in descending order, and only the top result is selected using LIMIT 1.



# Execute the query
cur.execute(query)

# Fetch all rows from the result set
result=cur.fetchall()

#Iterate over the result set and print the director with the most movies and the count of movies
for row in result:
    print('Director made most movies are:',row[0],',the count of movies are:',row[1])
```

### 13.Whose director is the most bankable

```python
# SQL query
query="Select name from directors D inner join Movies M on D.id = M.director_id group by M.director_id order by sum(budget) desc limit 1"
#This SQL query joins the directors and Movies tables on the id and director_id columns, respectively. 
#It then groups the result by the director's ID, calculates the sum of the budget column for each director, orders the result in descending order based on the sum of the budget, and selects only the top result using LIMIT 1.


# Execute the query
cur.execute(query)

# Fetch all rows from the result set
result=cur.fetchall()

# Iterate over the result set and print the name of the most bankable director
for row in result:
    print("The Most bankable director is:",result[0])
```

# Budget Analysis

## List out Top 10 highest budget made movies

```python
# SQL query
query="SELECT M.original_title,D.name,D.id,M.release_date,M.budget,M.revenue from movies M inner join directors D on M.director_id=D.id   order by budget desc limit 10"
#This SQL query joins the movies and directors tables on the director_id and id columns, respectively. 
#It selects specific columns (original_title, name, id, release_date, budget, and revenue) from the joined tables, orders the result by the budget column in descending order, and limits the result to the top 10 rows.

# Execute the query
cur.execute(query)
# Fetch all rows from the result set
result=cur.fetchall()
# Create a DataFrame from the result using pandas
mostExpensive = pd.DataFrame(result, columns = ['original_title', 'director_name','director_id','release_date', 'budget', 'revenue'])
# Display the top 10 rows of the DataFrame
mostExpensive.head(10)
```
